## 2. Przygotowanie danych

```python
import pandas as pd

hap = pd.read_csv(r'C:\Users\Gosia\Desktop\Python\dane\2016.csv') # szczęście
su = pd.read_csv(r'C:\Users\Gosia\Desktop\Python\dane\master.csv') # samobójstwa
```

```python
#Tabela szczęście
pd.set_option("display.max_rows", 15, "display.max_columns", None)
display(hap)
```

```python
#Tabela samobójstwa
display(su)
```

```python
#Zawężam dane do 2016 roku w tabeli su
su2 = su[(su['year'] == 2016 )]
display(su2)
```

```python
#Poprawiam nazwę kolumny i używam jej by połączyć obie tabele
suN = su2.rename(columns = {'country':'Country'}, inplace = False)
print('tabela samobójstw:')
print('')
print(suN.columns)
print('')
print('- - - - - - - - -')
print('tabela szczęścia:')
print('')
print(hap.columns)
```

```python
df = pd.merge(hap, suN, on='Country', how="inner")
df.head(5)
```

```python
#Sprawdzam ilość wierszy dla każdego kraju
df.groupby('Country').size()
```

```python
#Tworzę nową zmienną - zbiorczą ilość samobójstw dla danego kraju dla wszystkich grup wiekowych i płci
df['suicide_avg/100k'] = df['suicides/100k pop'].groupby(df['Country']).transform('sum')/10
df.head(11)
```

```python
#Usuwam kolumny, które na pewno się nie przydadzą
df2 = df.drop(columns=["Region", "sex", "age", "country-year", "HDI for year", "generation", "year", "suicides_no", "population", "suicides/100k pop"])
```

## 4. Prosty model regresji liniowej

<span style="color: blue;">
    Ze względu na bardzo niewielką ilość danych <b>(15 obserwacji)</b> nie dzielę zbioru na zbiór testowy i treningowy (ani walidacyjny). <br>Wykonuje operacje wyłącznie na pełnym zbiorze wejściowym.

```python
from statsmodels.formula.api import ols
```

```python
print(df2.columns)
```

```python
#Zmieniam nazwy kolumn, by łatwiej było utworzyć równanie w regresji
df3 = df2.rename(columns={'Happiness Rank': 'HappinessRank', 'Economy (GDP per Capita)': 'Economy', 'Health (Life Expectancy)': "Health",
                    'Trust (Government Corruption)': 'Trust', 'suicide_avg/100k': 'Suicides'}, inplace=False)
df3.head(3)
```

```python
#Sprawdzam czy wszystkie zmienne są zmiennymi ciągłymi
#df3.dtypes
```

```python
#Buduję model regresji wielorakiej na podstawie zadanych zmiennych objaśniających
model = ols('Suicides ~ Economy + Family + Health + Freedom + Trust', data=df3).fit()
print(model.summary())  
```

<span style="color: blue;">
    Maksymalne przeskalowane R-kwadrat wskazuje, że <b>41,5%</b> zmiennej Suicides zostało wyjaśnionych przez model.

<span style="color: blue;">Równanie regresji wielorakiej:<br>
<b>y = -14.1931 + 18.3549 * Economy + 34.8005 * Family - 9.1976 * Health - 44.9562 * Freedom - 19.7543 * Trust </b>

```python
#Przepisuje ręcznie równanie, bo z jakiegoś powodu kod poniżej się uruchamia, ale liczy coś w nieskończoność... :)
df3['Predict_Suicides'] = -14.1931 + 18.3549 * df3['Economy'] + 34.8005 * df3['Family'] - 9.1976 * df3['Health'] - 44.9562 * df3['Freedom'] - 19.7543 * df3['Trust']
df3.head(5)
```

```python
#predictors = ['Economy', 'Family', 'Health', 'Freedom', 'Trust']
#outcome_pred = model.predict(predictors)
```

```python
#Tabela do porównania prawdziwych wyników z prognozami 
compare = pd.DataFrame({'Country': df3['Country'], 'Actual Suicides': df3['Suicides'], 'Predicted Suicides': df3['Predict_Suicides']})
compare
```

```python
#Wykres wartości rzeczywistych w porównaniu z prognozowanymi
import matplotlib.pyplot as plt
data = compare.head(25)
data.plot(x="Country", kind='bar',figsize=(12,6), fontsize=12, color=['lightblue', 'orange'])
plt.grid(which='major', linewidth='0.5')
plt.grid(which='minor', linewidth='0.5')
plt.title('Actual vs Predicted', fontsize=12)
plt.xlabel("Countries", fontsize=12)
plt.ylabel("Suicides / 100k population", fontsize=12)
plt.show()
```

<span style="color: blue;">
    Na podstawie wcześniejszych statystyk i powyższego zestawienia oraz wykresu możemy stwierdzić, że taki model słabo odwzorowuje rzeczywistość i nie mógłby zostać użyty w prawdziwym życiu. Zmienne Economy, Family, Health, Freedom i Trust nie opisują dobrze zmiennej objaśnianej. <span>
